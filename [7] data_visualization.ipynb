```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Titanic.csv')

df['survived'] = df['survived'].astype(str)
df['class'] = df['class'].astype(str)

df
```

```python
# Bar Chart
sns.countplot(x='sex', hue='survived', data=df)
plt.title('Survival Rates by Gender')
plt.ylabel('Number of Passengers')
plt.xlabel('Gender')
plt.show()
```

```python
# Pie Chart Values
class_counts = df['class'].value_counts
class_counts.index
```

```python
# Pie Charts
plt.pie(class_counts, labels=class_counts.index, autopct = '%1.1f%%', startangle=90, colors=['gold', 'lightblue', 'lightgreen'])
plt.title('Passenger Class Distribution')
plt.show()
```

```python
import seaborn as sns

flights = sns.load_dataset('flights')
flights['year'] = flights['year'].astype(int)
flights
```

```python
# grouped and summed by year, row indexes are reset
yearly_passengers = flights.groupby('year')['passengers'].sum().reset_index()
yearly_passengers
```

```python
# Line chart with Seaborn
sns.lineplot(data=yearly_passengers, x='year', y='passengers')
plt.title('Total Passengers per Year')
plt.xlabel('Year')
plt.ylabel('Total Passengers')
plt.show()
```

```python
sns.lineplot(data=flights, x='month', y='passengers', hue='year', palette='viridis', marker='o')
plt.title('Monthly Passenger Numbers Over Years')
plt.xlabel('Month')
plt.ylabel('Number of passengers')
plt.legend(title='Year', bbox_to_anchor=(1.05,1), loc='upper left')
plt.show()
```

```python
# Scatter Plot Chart
sns.scatterplot(x='age', y='fare', hue='survived', style='alone', data=df)
plt.title('Scatter Plot: Age vs. Fare by Survival')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
# Regression Plot
sns.regplot(x='age', y='fare', data=df)
plt.title('Regression Plot: Fare vs. Age')
plt.xlabel('age')
plt.ylabel('Fare')
plt.show()

# y = mx + b
# fare = age * m + b
```

```python
fig, ax = plt.subplots(2, 2, figsize=(16, 12))

sns.countplot(ax=ax[0, 0], x='class', hue='survived', data=df)
ax[0, 0].set_title("Survival Count By Class")
sns.histplot(ax=ax[0, 1], data=df, x='age', hue='survived', multiple='stack')
ax[0, 1].set_title("Age Distribution by Survival")
sns.histplot(ax=ax[1, 0], data=df, x='age', y='fare', hue='survived', data=df)
ax[1, 0].set_title("Age vs. Fare by Survival")

plt.tight_layout()
plt.show()
```
